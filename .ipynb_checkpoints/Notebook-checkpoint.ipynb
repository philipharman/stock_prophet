```python
# Imports
import pandas as pd
import yfinance as yf
from prophet import Prophet
from prophet.plot import plot_plotly, plot_components_plotly
```

```python
# Specify stock symbol + setup df   
TICKER = 'AAPL' # Must be a valid selection

# get historical market data as df
stock = yf.Ticker(TICKER)
hist = pd.DataFrame(stock.history(period='3y'))

# Change date column to datetime type
hist.index = pd.to_datetime(hist.index)
hist.dropna(inplace=True)

pseries = pd.DataFrame()
pseries['ds'] = hist.index
pseries['y'] = hist.Close.values

model = Prophet()
model.fit(pseries)
future = model.make_future_dataframe(periods=180)
forecast = model.predict(future)
```

```python
fig = plot_plotly(model, forecast)
fig.update_layout(
    title = '6-Month Projection: {}'.format(TICKER),
    xaxis_title="Date",
    yaxis_title="Closing Value",
    font=dict(family="Courier New, monospace",size=18))
fig
```
